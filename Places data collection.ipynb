## Generate ikea list in different countries

```python
country_list = ['the Netherlands', 'the UK', 'France', 'Germany', 'Japan', 'Taiwan', 'the US']
country_list2 = ['Sweden', 'Hong Kong', 'Turkey', 'Australia', 'Canada', 'Spain', 'Italy', 'Belgium', 'Poland']
```

```python
ikea_list = []
for country in country_list2:
    query = 'ikea in '+ country
    query_result = gmaps.places(query=query)
    for item in query_result['results']:
        item['country'] = country
    ikea_list += query_result['results']
```

```python
ikea_df = pd.DataFrame.from_records(ikea_list)
ikea_df.to_csv("data/ikea_df2.csv", index=False)
ikea_df
```

## Generate places near each ikea

```python
type_list = ['cafe', 'restaurant', 'bar', 'clothing_store']
places_list = []
for index, ikea in ikea_df.iterrows():
    print(index, ikea['formatted_address'])
    ikea_location = ikea['geometry']['location']
    ikea_address = ikea['formatted_address']
    
    for place_type in type_list:
        query_result1 = gmaps.places_nearby(location=ikea_location, type=place_type, rank_by='distance')
        places_result = query_result1['results']
        if 'next_page_token' in query_result1:
            next_page_token = query_result1['next_page_token']
            time.sleep(3)
            query_result2 = gmaps.places_nearby(location=ikea_location, type=place_type, rank_by='distance', page_token=next_page_token)
            places_result += query_result2['results']
        for place in places_result:
            place['country'] = ikea['country']
            place['type'] = place_type
            place['ikea_location'] = ikea_location
            place['ikea_address'] = ikea_address
            place['ikea_place_id'] = ikea['place_id']
            place['ikea_rating'] = ikea['rating']
            place['ikea_user_ratings_total'] = ikea['user_ratings_total']
        places_list += places_result
    
```

```python
places_40_df = pd.DataFrame.from_records(places_list)
places_40_df = places_40_df.drop_duplicates(['place_id'])
places_40_df.to_csv('data/places_40_df2.csv', index=False)
places_40_df
```

```python
places_40_df['distance_to_ikea'] = 0
places_40_df['duration_to_ikea'] = 0 
ikea_location_list = ikea_df['geometry'].tolist()
ikea_location_list = [i['location'] for i in ikea_location_list]

for ikea_location in ikea_location_list:
    print(ikea_location)
    #seperate as it exceeded the limits
    for place_type in type_list:
        
        near_places_locations = places_40_df[(places_40_df['ikea_location']== ikea_location) & (places_40_df['type']==place_type)]['geometry'].tolist()
        near_places_locations = [i['location'] for i in near_places_locations]
        if len(near_places_locations)>0:
            distance_matrix = gmaps.distance_matrix(origins=near_places_locations, destinations=ikea_location)

            distances_list = [i['elements'][0]['distance']['value'] for i in distance_matrix['rows']]
            durations_list = [i['elements'][0]['duration']['value'] for i in distance_matrix['rows']]

            places_40_df.loc[(places_40_df['ikea_location']== ikea_location) & (places_40_df['type']==place_type), 'distance_to_ikea'] = distances_list
            places_40_df.loc[(places_40_df['ikea_location']== ikea_location) & (places_40_df['type']==place_type), 'duration_to_ikea'] = durations_list
```

```python
places_40_df.to_csv('data/places_include_distance2.csv', index=False)
places_40_df
```
